```python
"""
StockVision — Phase 1: Data Exploration (v2)
========================================
Angepasst für LSTM v4:
- Alle absoluten Preis-Features werden in stationäre relative Features umgewandelt
- Neue Feature-Visualisierungen
- Gleiche stationäre Features wie im LSTM v4 Training
"""

# ==========================================
# Imports
# ==========================================
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import time
from pathlib import Path

pd.set_option("display.max_columns", None)
sns.set(style="whitegrid")

# ==========================================
# Setup Paths
# ==========================================
if os.path.basename(os.getcwd()) == "notebooks":
    DATA_RAW       = Path("../data/raw")
    DATA_PROCESSED = Path("../data/processed")
else:
    DATA_RAW       = Path("data/raw")
    DATA_PROCESSED = Path("data/processed")

DATA_RAW.mkdir(parents=True, exist_ok=True)
DATA_PROCESSED.mkdir(parents=True, exist_ok=True)

print("Data folders ready:")
print(f"  Raw:       {DATA_RAW}")
print(f"  Processed: {DATA_PROCESSED}")

# ==========================================
# Load Data
# ==========================================
TICKER     = "AAPL"
START_DATE = "2015-01-01"
END_DATE   = "2025-01-01"

print(f"\nLoading {TICKER} ({START_DATE} → {END_DATE})...")


def fetch_history(ticker_symbol, start, end, retries=3, sleep_s=2.0):
    last_err = None
    for attempt in range(1, retries + 1):
        try:
            t  = yf.Ticker(ticker_symbol)
            df = t.history(start=start, end=end, auto_adjust=False)
            if df is not None and len(df) > 0:
                if isinstance(df.index, pd.DatetimeIndex) and df.index.tz is not None:
                    df.index = df.index.tz_localize(None)
                return df
            last_err = RuntimeError("yfinance returned empty DataFrame.")
        except Exception as e:
            last_err = e
        print(f"  Attempt {attempt}/{retries} failed: {last_err}")
        if attempt < retries:
            time.sleep(sleep_s)
    raise RuntimeError(f"Could not fetch '{ticker_symbol}'. Last error: {last_err}")


data = fetch_history(TICKER, START_DATE, END_DATE, retries=4)
print(f"Loaded: {data.shape} | {data.index[0].date()} → {data.index[-1].date()}")

# Rohdaten speichern
raw_path = DATA_RAW / f"{TICKER}_raw.csv"
data.to_csv(raw_path)
print(f"Raw data saved: {raw_path}")

# ==========================================
# Basic Inspection
# ==========================================
print("\n" + "="*60)
print("DATA INSPECTION")
print("="*60)
data.info()
print("\nStatistical Summary:")
display(data.describe())

missing = data.isnull().sum()
if missing.sum() == 0:
    print("\nNo missing values ✅")
else:
    print("\nMissing values:")
    display(missing[missing > 0])

# ==========================================
# Price & Volume Plot
# ==========================================
fig, axes = plt.subplots(2, 1, figsize=(14, 8))
fig.suptitle(f"{TICKER} Raw Data (2015–2025)", fontsize=16, fontweight="bold")

axes[0].plot(data.index, data["Close"], linewidth=1.5, label="Close Price")
axes[0].set_ylabel("Price (USD)")
axes[0].set_title("Closing Price")
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].bar(data.index, data["Volume"], width=1, alpha=0.6, label="Volume")
axes[1].set_ylabel("Volume")
axes[1].set_title("Trading Volume")
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ==========================================
# Feature Engineering — Basis-Indikatoren
# (werden für stationäre Features gebraucht)
# ==========================================
print("\n" + "="*60)
print("FEATURE ENGINEERING — BASISINDIKATOREN")
print("="*60)

data["Return"]       = data["Close"].pct_change()
data["MA7"]          = data["Close"].rolling(7).mean()
data["MA30"]         = data["Close"].rolling(30).mean()
data["MA90"]         = data["Close"].rolling(90).mean()
data["Volatility"]   = data["Return"].rolling(30).std()
data["Volume_MA20"]  = data["Volume"].rolling(20).mean()

delta = data["Close"].diff()
gain  = delta.where(delta > 0, 0).rolling(14).mean()
loss  = (-delta.where(delta < 0, 0)).rolling(14).mean()
data["RSI"] = 100 - (100 / (1 + gain / loss))

data["HL_Spread"] = (data["High"] - data["Low"]) / data["Close"]
data["Price_Change"] = data["Close"].diff()

ema12 = data["Close"].ewm(span=12, adjust=False).mean()
ema26 = data["Close"].ewm(span=26, adjust=False).mean()
data["MACD"]        = ema12 - ema26
data["MACD_Signal"] = data["MACD"].ewm(span=9, adjust=False).mean()

ma20  = data["Close"].rolling(20).mean()
std20 = data["Close"].rolling(20).std()
data["BB_Upper"] = ma20 + 2 * std20
data["BB_Lower"] = ma20 - 2 * std20

hl  = data["High"] - data["Low"]
hc  = (data["High"] - data["Close"].shift()).abs()
lc  = (data["Low"]  - data["Close"].shift()).abs()
data["ATR"] = pd.concat([hl, hc, lc], axis=1).max(axis=1).rolling(14).mean()

data["Momentum_5"]  = data["Close"].diff(5)
data["Momentum_10"] = data["Close"].diff(10)

low14  = data["Low"].rolling(14).min()
high14 = data["High"].rolling(14).max()
data["Stochastic"] = 100 * (data["Close"] - low14) / (high14 - low14)

data = data.dropna().copy()
print(f"Nach Basis-Indikatoren + dropna: {len(data)} Rows")

# ==========================================
# Stationäre Features (identisch mit LSTM v4)
# WICHTIG: Diese Features gehen ins Modell, nicht die absoluten Werte
# ==========================================
print("\n" + "="*60)
print("STATIONÄRE FEATURES BAUEN (wie LSTM v4)")
print("="*60)

feat = pd.DataFrame(index=data.index)

# Returns
feat["return_1d"]   = data["Close"].pct_change()
feat["return_3d"]   = data["Close"].pct_change(3)
feat["return_5d"]   = data["Close"].pct_change(5)
feat["return_10d"]  = data["Close"].pct_change(10)
feat["return_21d"]  = data["Close"].pct_change(21)

# MA-Abstände: wie weit ist der Preis von der MA entfernt?
feat["dist_ma7"]    = data["Close"] / data["MA7"]  - 1
feat["dist_ma30"]   = data["Close"] / data["MA30"] - 1
feat["dist_ma90"]   = data["Close"] / data["MA90"] - 1
feat["ma7_vs_ma30"] = data["MA7"]   / data["MA30"] - 1  # Crossover-Signal

# Bollinger Bands: Position im Band (0=unten, 1=oben)
bb_range = (data["BB_Upper"] - data["BB_Lower"]).replace(0, np.nan)
feat["bb_position"] = (data["Close"] - data["BB_Lower"]) / bb_range
feat["bb_width"]    = (data["BB_Upper"] - data["BB_Lower"]) / data["Close"]

# MACD normiert auf Preis
feat["macd_norm"]        = data["MACD"] / data["Close"]
feat["macd_signal_norm"] = data["MACD_Signal"] / data["Close"]
feat["macd_hist"]        = (data["MACD"] - data["MACD_Signal"]) / data["Close"]

# RSI auf [0,1]
feat["rsi"]         = data["RSI"] / 100.0

# Volatilität (bereits als Return-Std, stationär)
feat["volatility"]  = data["Volatility"]

# Volume relativ zur MA
feat["volume_ratio"] = data["Volume"] / data["Volume_MA20"] - 1

# ATR normiert
feat["atr_norm"]    = data["ATR"] / data["Close"]

# HL-Spread (bereits relativ)
feat["hl_spread"]   = data["HL_Spread"]

# Stochastic auf [0,1]
feat["stochastic"]  = data["Stochastic"] / 100.0

# Momentum normiert
feat["mom_5_norm"]  = data["Momentum_5"]  / data["Close"]
feat["mom_10_norm"] = data["Momentum_10"] / data["Close"]

feature_cols = list(feat.columns)
print(f"\nStationäre Features ({len(feature_cols)}):")
for f in feature_cols:
    print(f"  {f}")

# Prüfe Stationarität: Mittelwerte sollten nahe 0 sein
print("\nStationaritäts-Check (Mittelwerte, sollten alle nahe 0 sein):")
means = feat.mean()
for col, val in means.items():
    flag = "⚠️  " if abs(val) > 1 else "✅"
    print(f"  {flag} {col}: {val:.6f}")

# ==========================================
# Targets
# ==========================================
feat["Close"]        = data["Close"]
feat["Next_Close"]   = data["Close"].shift(-1)
feat["Return_Next"]  = data["Close"].pct_change().shift(-1)

threshold = 0.005
feat["Direction_Binary"] = (feat["Next_Close"] > feat["Close"]).astype(int)
feat["Direction_3Class"] = pd.cut(
    feat["Return_Next"],
    bins=[-np.inf, -threshold, threshold, np.inf],
    labels=[0, 1, 2],
).astype(float)

feat = feat.replace([np.inf, -np.inf], np.nan).dropna().copy()
print(f"\nFinales Dataset: {len(feat)} Rows, {len(feat.columns)} Columns")

# ==========================================
# Distributions der stationären Features
# ==========================================
print("\nVisualisiere stationäre Feature-Verteilungen...")

n_cols = 4
n_rows = int(np.ceil(len(feature_cols) / n_cols))
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, n_rows * 3))
axes = axes.flatten()

for i, col in enumerate(feature_cols):
    axes[i].hist(feat[col].dropna(), bins=50, alpha=0.7, edgecolor="black", linewidth=0.3)
    axes[i].axvline(0, color="red", linewidth=1, linestyle="--", alpha=0.7)
    axes[i].set_title(col, fontsize=9)
    axes[i].set_ylabel("Freq", fontsize=8)

for j in range(len(feature_cols), len(axes)):
    axes[j].set_visible(False)

fig.suptitle("Stationäre Feature-Verteilungen (sollten um 0 zentriert sein)", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.show()

# ==========================================
# Zeitreihen-Plots der wichtigsten Features
# ==========================================
fig, axes = plt.subplots(4, 2, figsize=(15, 14))
fig.suptitle(f"{TICKER} — Stationäre Features (Zeitreihen)", fontsize=14, fontweight="bold")

plot_pairs = [
    ("dist_ma7",    "dist_ma30",   "MA-Abstände (Close/MA - 1)"),
    ("bb_position", "bb_width",    "Bollinger Band Position & Breite"),
    ("macd_norm",   "macd_hist",   "MACD normiert"),
    ("rsi",         "stochastic",  "RSI & Stochastic (auf [0,1])"),
    ("return_1d",   "return_5d",   "Returns 1d & 5d"),
    ("volatility",  "atr_norm",    "Volatilität & ATR normiert"),
    ("volume_ratio","hl_spread",   "Volume Ratio & HL-Spread"),
    ("mom_5_norm",  "mom_10_norm", "Momentum 5d & 10d (normiert)"),
]

for ax, (col1, col2, title) in zip(axes.flatten(), plot_pairs):
    ax.plot(feat.index, feat[col1], linewidth=0.8, label=col1, alpha=0.8)
    ax.plot(feat.index, feat[col2], linewidth=0.8, label=col2, alpha=0.8)
    ax.axhline(0, color="black", linewidth=0.6, linestyle="--")
    ax.set_title(title, fontsize=9)
    ax.legend(fontsize=7)
    ax.grid(True, alpha=0.2)

plt.tight_layout()
plt.show()

# ==========================================
# Korrelation der stationären Features
# ==========================================
print("\nKorrelationsmatrix der stationären Features...")

corr = feat[feature_cols].corr()

plt.figure(figsize=(14, 12))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(
    corr,
    mask=mask,
    annot=True,
    fmt=".2f",
    square=True,
    linewidths=0.3,
    cmap="RdBu_r",
    center=0,
    cbar_kws={"shrink": 0.8},
    annot_kws={"size": 7},
)
plt.title(f"{TICKER} — Feature-Korrelationen (stationär)", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.show()

# Hochkorrelierte Feature-Paare ausgeben
print("\nHochkorrelierte Feature-Paare (|r| > 0.8):")
high_corr = []
for i in range(len(corr.columns)):
    for j in range(i + 1, len(corr.columns)):
        r = corr.iloc[i, j]
        if abs(r) > 0.8:
            high_corr.append((corr.columns[i], corr.columns[j], round(r, 3)))

if high_corr:
    for a, b, r in sorted(high_corr, key=lambda x: abs(x[2]), reverse=True):
        print(f"  {a} ↔ {b}: {r}")
else:
    print("  Keine stark korrelierten Features ✅")

# ==========================================
# Target Distributions
# ==========================================
print("\n" + "="*60)
print("TARGET DISTRIBUTIONS")
print("="*60)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Return_Next Distribution
axes[0].hist(feat["Return_Next"], bins=60, alpha=0.7, edgecolor="black", linewidth=0.3)
axes[0].axvline(0, color="red", linewidth=1.5, linestyle="--")
axes[0].set_title("Return_Next Distribution (LSTM Target)")
axes[0].set_xlabel("Return_Next")
axes[0].set_ylabel("Frequency")

# Direction Binary
dist_bin = feat["Direction_Binary"].value_counts(normalize=True).sort_index()
axes[1].bar(["Down (0)", "Up (1)"], dist_bin.values, color=["#e74c3c", "#2ecc71"], alpha=0.8)
axes[1].set_title("Direction Binary (XGBoost Target)")
axes[1].set_ylabel("Anteil")
for i, v in enumerate(dist_bin.values):
    axes[1].text(i, v + 0.005, f"{v:.1%}", ha="center", fontsize=11)

# Direction 3-Class
dist_3 = feat["Direction_3Class"].value_counts(normalize=True).sort_index()
axes[2].bar(["Down (0)", "Neutral (1)", "Up (2)"], dist_3.values,
            color=["#e74c3c", "#f39c12", "#2ecc71"], alpha=0.8)
axes[2].set_title("Direction 3-Class")
axes[2].set_ylabel("Anteil")
for i, v in enumerate(dist_3.values):
    axes[2].text(i, v + 0.005, f"{v:.1%}", ha="center", fontsize=11)

plt.tight_layout()
plt.show()

print(f"\nDirection Binary:  Down={dist_bin.get(0, 0):.1%}, Up={dist_bin.get(1, 0):.1%}")
print(f"Direction 3-Class: Down={dist_3.get(0.0, 0):.1%}, Neutral={dist_3.get(1.0, 0):.1%}, Up={dist_3.get(2.0, 0):.1%}")

# ==========================================
# Train/Val/Test Split (temporal, 70/15/15)
# ==========================================
print("\n" + "="*60)
print("TRAIN / VAL / TEST SPLIT")
print("="*60)

n = len(feat)
train_end = int(n * 0.70)
val_end   = int(n * 0.85)

train = feat.iloc[:train_end].copy()
val   = feat.iloc[train_end:val_end].copy()
test  = feat.iloc[val_end:].copy()

print(f"  Train: {len(train)} Rows  ({train.index[0].date()} → {train.index[-1].date()})")
print(f"  Val:   {len(val)} Rows  ({val.index[0].date()} → {val.index[-1].date()})")
print(f"  Test:  {len(test)} Rows  ({test.index[0].date()} → {test.index[-1].date()})")
print(f"  Ratio: {len(train)/n:.0%} / {len(val)/n:.0%} / {len(test)/n:.0%}")

# ==========================================
# Speichern
# ==========================================
print("\n" + "="*60)
print("SAVING")
print("="*60)

# Komplettes Dataset mit allen Features + Targets
full_path = DATA_PROCESSED / f"{TICKER}_features_stationary.csv"
feat.to_csv(full_path, index=True)
print(f"Full stationary dataset: {full_path}")

# Splits separat
train.to_csv(DATA_PROCESSED / f"{TICKER}_train.csv", index=True)
val.to_csv(DATA_PROCESSED / f"{TICKER}_val.csv",   index=True)
test.to_csv(DATA_PROCESSED / f"{TICKER}_test.csv",  index=True)
print(f"Train/Val/Test splits gespeichert.")

# feature_cols für spätere Notebooks
import json
meta = {
    "ticker":       TICKER,
    "feature_cols": feature_cols,
    "n_features":   len(feature_cols),
    "n_total":      len(feat),
    "date_range":   [str(feat.index.min().date()), str(feat.index.max().date())],
    "splits": {
        "train_end": train_end,
        "val_end":   val_end,
    }
}
meta_path = DATA_PROCESSED / f"{TICKER}_meta.json"
meta_path.write_text(json.dumps(meta, indent=2), encoding="utf-8")
print(f"Metadata gespeichert: {meta_path}")

# ==========================================
# Summary
# ==========================================
print("\n" + "="*60)
print("PHASE 1 COMPLETE ✅")
print("="*60)
print(f"  Ticker:           {TICKER}")
print(f"  Rows (clean):     {len(feat)}")
print(f"  Features:         {len(feature_cols)} stationäre Features")
print(f"  Targets:          Return_Next, Direction_Binary, Direction_3Class")
print(f"  Splits:           {len(train)} / {len(val)} / {len(test)}")
print(f"\nNächste Schritte:")
print(f"  Notebook 02: LSTM v4 mit stationären Features")
print(f"  Notebook 03: XGBoost Classifier")
print(f"\nLaden in Notebook 02:")
print(f"  feat = pd.read_csv(r'{full_path}', index_col=0, parse_dates=True)")
print(f"  meta = json.loads(Path(r'{meta_path}').read_text())")
print(f"  feature_cols = meta['feature_cols']")
```
